## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
popular_df.head()
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```

```python
books['Year-Of-Publication'] = pd.to_numeric(books['Year-Of-Publication'], errors='coerce')
books = books.dropna(subset=['Year-Of-Publication'])
books['Year-Of-Publication'] = books['Year-Of-Publication'].astype(int)
top_10_years = books['Year-Of-Publication'].value_counts().head(10)
books_top_10_years = books[books['Year-Of-Publication'].isin(top_10_years.index)]

plt.figure(figsize=(10, 6))
sns.histplot(data=books_top_10_years, x='Year-Of-Publication', kde=True, bins=10)
plt.title('Distribution of Books Published Over the Years (Top 10 Years)')
plt.xlabel('Year of Publication')
plt.ylabel('Frequency')
plt.show()
```

```python
top_10_books = books.sort_values('Year-Of-Publication', ascending=False).head(10)
merged_data = pd.merge(ratings, top_10_books, on='ISBN')
merged_data = pd.merge(merged_data, users, on='User-ID')
book_ratings = merged_data.groupby('Book-Title')['Book-Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(book_ratings.index, book_ratings)
plt.title('Average Ratings for Top 10 Books Published')
plt.xlabel('Book Title')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
merged_data = pd.merge(ratings, books, on='ISBN')

book_ratings = merged_data.groupby('Book-Title').agg({'Book-Rating': 'mean', 'ISBN': 'count'})
book_ratings = book_ratings.rename(columns={'ISBN': 'Rating-Count'})

plt.figure(figsize=(10, 6))
plt.scatter(book_ratings['Rating-Count'], book_ratings['Book-Rating'], alpha=0.5)
plt.title('Average Rating vs Rating Count')
plt.xlabel('Rating Count')
plt.ylabel('Average Rating')
plt.show()
```

```python
top_10_books = books.sort_values('Year-Of-Publication', ascending=False).head(10)

merged_data = pd.merge(ratings, top_10_books, on='ISBN')

book_ratings = merged_data.groupby('Book-Title').agg({'Book-Rating': 'mean', 'ISBN': 'count'})
book_ratings = book_ratings.rename(columns={'ISBN': 'Rating-Count'})

plt.figure(figsize=(10, 6))
plt.scatter(book_ratings['Rating-Count'], book_ratings['Book-Rating'], alpha=0.5)
plt.title('Average Rating vs Rating Count for Top 10 Books')
plt.xlabel('Rating Count')
plt.ylabel('Average Rating')
plt.show()
```
